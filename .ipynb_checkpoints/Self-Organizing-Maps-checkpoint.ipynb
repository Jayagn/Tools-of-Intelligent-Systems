- Importing required libraries

```python
import numpy as np
from sklearn import preprocessing
from matplotlib import pyplot as plt
from matplotlib import patches as patches
%matplotlib inline
```

```python
#randomly selecting 24 colours
raw_data = np.array([[255, 0, 0], [0, 255, 0], [0, 0, 255], [0, 255,255],[255, 0,255], [153,255, 204],[102,178, 255],[255,160,122],
                     [255, 204, 255],[229,255,204],[255,102,102],[0,128,255],[255,0,127],[204,102,0],[204,0,204],[192,192,192],
                    [153,255,153],[255,255,204],[229,204,255],[204,204,0],[128,0,128],[0,128,128],[0,0,128],[250,128,114]]).T
```

- Normalizing data in range of 0 to 1.

```python
data = preprocessing.normalize(raw_data)
```

- Creating output grid of 100 * 100 Dimensions.
- Initialized learning rate to 0.8

```python
network_dimensions = np.array([100, 100])
n_iterations = 1000
init_learning_rate = 0.8
# establish size variables based on data
m = raw_data.shape[0]
n = raw_data.shape[1]

# weight matrix (i.e. the SOM) needs to be one m-dimensional vector for each neuron in the SOM
net = np.random.random((network_dimensions[0], network_dimensions[1], m))
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')
ax.set_xlim((0, net.shape[0]+1))
ax.set_ylim((0, net.shape[1]+1))
ax.set_title('Self-Organising Map at random weights')

#plot the map
for x in range(1, net.shape[0] + 1):
    for y in range(1, net.shape[1] + 1):
        ax.add_patch(patches.Rectangle((x-0.5, y-0.5), 1, 1,facecolor=net[x-1,y-1,:],edgecolor='none'))
plt.show()
```

Applying Self Organizing Map Algorithm
- Finding the neuron in the SOM whose associated 3D vector is closest to our chosen 3D colour vector. The neuron obtained is winning node.
- Moving the winning node 3D weight vector closer to the input vector in 3D space.
- Identifying the 2D neighbours of the winning node and also move their 3D weight vectors closer to the input vector.
- Lastly, updating the learning rate.

```python
def find_winning_node(t, net, m):
    winning_node_idx = np.array([0, 0])
    # set the initial minimum distance to a huge number
    min_dist = np.iinfo(np.int).max
    # calculating high-dimensional distance between each neuron and the input
    for x in range(net.shape[0]):
        for y in range(net.shape[1]):
            w = net[x, y, :].reshape(m, 1)
            #Getting Squared Euclidean distance.
            sq_dist = np.sum((w - t) ** 2)
            if sq_dist < min_dist:
                min_dist = sq_dist
                bmu_idx = np.array([x, y])
    winning_node = net[winning_node_idx[0], winning_node_idx[1], :].reshape(m, 1)
    return (winning_node, winning_node_idx)
```

```python
def decay_radius(initial_radius, i, time_constant):
    return initial_radius * np.exp(-i / time_constant)

def decay_learning_rate(initial_learning_rate, i, n_iterations):
    return initial_learning_rate * np.exp(-i / n_iterations)
    
def calculate_influence(distance, radius):
    return np.exp(-distance / (2* (radius**2)))
```

```python
#'init' is values of sigma and 'it' is number of iterations 
init = [1,10,30,50,70]
it = [20, 40,100,1000]
count = 0
for j in range(len(init)):
  for i in range(n_iterations):
      # select a training example at random
      t = data[:, np.random.randint(0, n)].reshape(np.array([m, 1]))
      
      # find the winning node
      winning_node, winning_node_idx = find_winning_node(t, net, m)
      
      # decay the SOM parameters
      time_constant = n_iterations / np.log(init[j])
      r = decay_radius(init[j], i, time_constant)
      l = decay_learning_rate(init_learning_rate, i, n_iterations)
      
      # now we know the BMU, update its weight vector to move closer to input
      # and move its neighbours in 2-D space closer
      # by a factor proportional to their 2-D distance from the BMU
      for x in range(net.shape[0]):
          for y in range(net.shape[1]):
              w = net[x, y, :].reshape(m, 1)
              w_dist = np.sum((np.array([x, y]) - winning_node_idx) ** 2)
              # if the distance is within the current neighbourhood radius
              if w_dist <= r**2:
                  # calculate the degree of influence (based on the 2-D distance)
                  influence = calculate_influence(w_dist, r)
                  new_w = w + (l * influence * (t - w))
                  net[x, y, :] = new_w.reshape(1, 3)

      if(i == it[count]-1):
            fig = plt.figure()
            ax = fig.add_subplot(111, aspect='equal')
            ax.set_xlim((0, net.shape[0]+1))
            ax.set_ylim((0, net.shape[1]+1))
            ax.set_title('Self-Organising Map after {0} iterations and initial radius {1}'.format(it[count] ,init[j]))

            #plot the map
            for x in range(1, net.shape[0] + 1):
                for y in range(1, net.shape[1] + 1):
                    ax.add_patch(patches.Rectangle((x-0.5, y-0.5), 1, 1,facecolor=net[x-1,y-1,:],edgecolor='none'))
            plt.show()
            count+=1
  count=0
```

## Important notes

- From the above code cell 'init' is the list of initial radius(spread) and 'it' is the number of iterations.
- After every 20,40,100 and 1000 iterations the output figure is plotted, for each spread value that is (1,10,20,30,40,70) using matplotlib.

- The difference amoung all the outputs is clearly visible.
- The initial learning rate is set equal to 0.8.
- The decrease in learning rate is mandatory as in order to ensure convergence. In other words, if we do not change this value, learning procedure might not stop at all.
- The output is clearly not appropriate with spread equal to 1,10,20
- The output seems to be quite good, that is colours are clearly distinguished as the spread increases from 30. Also , as we increase number of iterations the output seems to be good.
- A good well converged map with several clusters is obtained after 1000 iterations and spread equal to 70.
- Hence, with increase in spread and iteration the map starts to converge well.
